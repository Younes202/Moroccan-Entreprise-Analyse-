## Import/Wrangling

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import re
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge  # noqa F401
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import make_pipeline
```

```python
# function for handling & cleaning our data
def wrangle(url):
    
    # import the file into a dataframe
    df = pd.read_csv(url)
    
    
    # Remove duplicate rows and keep only the first occurrence
    df.drop_duplicates(inplace=True)

    # Drop leaky columns 
    df.drop(columns=["email","web-site","fax","phone","phone2","schedules"], inplace=True)
    
    # Drop rows with empty description 
    df.dropna(subset=["description"], inplace=True)

    # Import moroccan cites json file
    file_path = "data/moroccan_cities.json"

    # Open the JSON file and load the data
    with open(file_path, "r") as json_file:
        moroccan_cities = json.load(json_file)

    # Based on the description feature retreive cites
    def check_city_mention(description):
        mentioned_cities = []  # List to store mentioned cities
        if isinstance(description, str):  # Check if it's a string
            for city in moroccan_cities:
                if city.lower() in description.lower():
                    mentioned_cities.append(city)  # Add mentioned city to the list
        return ', '.join(mentioned_cities) if mentioned_cities else None     # Return a string containing all mentioned cities separated by commas, or None if no city is mentioned

    # Add cites feature based on the description feature 
    df["cities"] = df["description"].apply(check_city_mention)
    # Add cites_na feature based on the name feature 
    df["cities_na"] = df["name"].apply(check_city_mention)
    
    # Replace NaN values in "cities" with values from "cities_na"
    df.loc[df["cities"].isna() & ~df["cities_na"].isna(), "cities"] = df["cities_na"]

    # Drop cities_na row
    df.drop(columns=["cities_na"], inplace=True)
    
    with open("data/sector_keywords.json", "r") as json_file:
        sector_keywords = json.load(json_file)

    # Function to search for sector keywords in description
    def search_sector_keywords(description):
        matches = []
        for sector, keywords in sector_keywords.items():
            for keyword in keywords:
                # Search for keyword in description (case insensitive)
                if re.search(r'\b{}\b'.format(keyword), description, re.IGNORECASE):
                    matches.append(sector)
                    break  # Break out of inner loop if keyword is found
        return matches
    
    df["sector"] = df["description"].apply(search_sector_keywords)

    return df
```

```python
url = "data/moroccan-entreprise-origin.csv"
df = wrangle(url)
df.to_csv("data/moroccan-entreprise-cleaned.csv", index=False)
```
